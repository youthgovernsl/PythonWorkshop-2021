# Type casting 

Casting is used when we need to specify a type to a variable.

Type casting can be categorized as



1.   Implicit Type Conversion - When the interpreter perform the type conversion without programmer's intervention.
2.   Explicit Type Conversion - When users convert the implicit data type of an object to the required data type.

In this section we will discuss about explicit type conversion using pre-defined in-built functions.

### Casting to int type 

```python
print(int(1))
print(int(3.8))
print(int(3.2))
print(int("3"))
```

###Logical Operation

'and' operation : Prints true if both the statements are true

```python

a,b = 5,7
print(a>=5 and b>=3)
print(a>5 and b>=3)
```

'or' operation : Prints true if atleast one of the statements are true.

```python

a,b = 5,7
print(a>=5 or b>=3)
print(a>5 or b>=3)
print(a>5 or b<3)
```

'not' operation : Reverse the result

```python

a,b = 5,7
print(not(a>=5 or b>=3))
```

###Identity Operators




'is' operator : Returns true if both variables are the same object

```python


x = ["hello", "world"]
y = ["hello","world"]
z = x
print(x==y)
print(x is y) #They have the same content, but its not the same object
print(x is z) #x and z is the same object
```

```python
x=["hello", "world"]
y="hello"
z="python"
print(y in x)
print(z not in x)
```

# Python Classes/Objects

Python supports object oriented programming. Hence, almost everything in Python 

1.   List item
2.   List item

is an object.

```python
#creating a class using the keyword 'class'
class MyClass:
  greeting = "Hello World!"
```

```python
#creating an object for the class created above
sentence = MyClass()
print(sentence.greeting)
```

```python
#creating a class using the __init__() built-in fucntion
class SemesterGrades:
  def __init__(student, name, grade):
    student.name = name
    student.grade = grade 

student1 = SemesterGrades("Samantha", "B+")

print(student1.name)
print(student1.grade)
```

```python
#creating a class with __init__() function and another user defined function
class vehicles:
  def __init__(car, model, year):
    car.model = model
    car.year = year
  
  #user defined fucntion
  def myFunct(carName):
    print("This is a/an", carName.model, "which was built in", carName.year)

car1 = vehicles("Audi", 2020)
car2 = vehicles("Acura", 2018)
car1.myFunct()
print(car2.model)  #This will only print the model of the car
```
