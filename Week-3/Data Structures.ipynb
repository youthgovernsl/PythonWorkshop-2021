#<u> Lists </u>


* Lists are the most versatile (adoptable, useful) of Python's
compound data types. A list contains items separated by
commas and enclosed within square brackets ([]).
* These are usually used to store multiple data
* The type of data in the list can be vary
* The items in the list are changable and duplicate values are allowed



```
student = ['Sunil',25,6.5,40,'school lane Bambalapitiya',718106745]
numbers = [1,2.01,3,4.32,5,6.2,7.89]
print(student)
print(numbers)
```


* Note that the above code is used to store data about a student and data of 
different data types is stored within the same list
* The name of the list is "student" and the data about that student is given enclosed with '[]' seperated by comas. This is the syntax of a python list
*  The second list consists of integers and floats

## Accessing elements
We can print all the elements in the list using the following command

```
print(student)
```

But if we want to access the information within the list individually we can do that by using indexes.
The elements in a list are indexed starting from 0,

*   Python provides a convenient way of taking parts of a list by
allowing us to denote a specific range of elements using the
colon symbol (:).
* There are multiple ways in which we can perform this
colon-based slicing operation.


```
print("First element in the list:" ,student[0])

# To slice a sequence of elements from a specified index value to the end, use [Index:]
print("Elements from index 2 onwards:" ,student[2:])

#  To slice a sequence of elements from beginning to a range denoted by an index value, use [:Index]
print("Elements upto the 4th index:" ,student[:4])

# To slice a sequence of elements within a range, use [Start Index:End Index]
print("Elements from index 2 upto the 4th index:" ,student[2:4])

# To reverse the whole list with the use of slicing operation, use [::-1]
print("Reversing the list:" ,student[::-1])
```

Note that when it is said upto the 4th index that the 4th index is not included.

Suppose that the student in the above list decided to change his phone number. We need to update the list. is can be done as follows,
We have to access the index with the phone number and change it to the new value.

```
student[5] = 771233214
print(student)
```

## Multi dimensional lists
Multi dimensional lists are an important part of programming. Suppose a table with rows and columns. The list that we have learned would not suffice to manipulate such data. For that a 2D list is used.

Consider the following example,
2 shops sell carrots and potatoes at different prices as bellow

Shop 1
*   carrot - 60
*   potato - 70


Shop 2
*   carrot - 55
*   potato - 75

Represent these in a 2D array and calculate the price of 1kg carrot and 1kg potato in each shop.

```
shopList = [[60,90],[55,75]]
for x in shopList:
  print("The prices of vegetables in shop",shopList.index(x)+1,"is :",x)

print("------------------------------------------------")
for x in shopList:
  tot = 0
  for y in x:
    tot += y
  print("The sum of prices in shop",shopList.index(x)+1,"is :",tot)
```

We can also use 3d arrays.Suppose an example of a 3,4,6 3D array whose each element is *

```
list3d = [[ ['*' for col in range(6)] for col in range(4)] for row in range(3)]
print(list3d)
```

#<u> Tuples </u>


*   Tuples are similar to lists. Howerver, once the values are assigned to a tuple , these cannot be modified during the run time
* Similar to lists , the values in the tuple can be accesed by their respectve indexes.


*   A tuple is defined by giving the values seperated by commas, enclosed with "()"
*   Duplicate items withn a lcould be present within a tuple.

```
month=('January','January','February','March','April','May','June','July','August','September','October','November',' December')
print("1:",month) #printing the whole tupple/Notice that the values can be duplcated
print("2:",month[0])
print("3:",month[1:3])
print("4:",month[2:])
print("5:",month*2)
```

We can join 2 tuples as follows. A tupple can also contain elements of various data types.

```
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

Now let us try to reassign the values to an already declared tuple.

```
fruits=("Apple","Oranges","Pineapple")
print("1:",fruits)
print("2:",fruits[0])
fruits[0] = "Mango" # we cannot assign the value
```

We can also try to see whether an item exists within a trupple using an if condition

```
fruits=("Apple","Oranges","Pineapple")
if "Banana" in fruits:
  print("Yes, 'Banana' is in the fruits tuple")
elif "Apple" in fruits:
  print("Yes, 'Apple' is in the fruits tuple")
```

## Updating values in a tuple


*   As mentioned earlier , we cannot update the values in a tuple. In other words , a tuple is immutable
*  However, a tuple can be converted to a list and vise versa


*   So we are able to convert the tuple into a list, perform all the operations needed and convert it back to a tuple.

```
fruits = ("apple", "banana", "cherry")
print("Before updating :",fruits)
y = list(fruits)
y[1] = "kiwi"
fruits = tuple(y)

print("After updating :",fruits)
```

We cannot append values to a tupple but we can append to a list

```
fruits = ("apple", "banana", "cherry")
fruits.append("Kiwi")
```

```
fruits = ("apple", "banana", "cherry")
print("Before appending :",fruits)
y = list(fruits)

y.append("kiwi") #adding values to the list
y.append("Mango")

y.remove("apple") #removing the value from the list

fruits = tuple(y)

print("After appending :",fruits)
```

#<u> Dictionary </u>



*   A dictionary is used to store values by the means of key value pairs
*   The key is used to access the value of the elements
*   Dictionaries are written withn curly braces with the first element as a key defined within " " and the value as the second element
*   Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.
*   Dictionaries are unordered, it means that the items does not have a defined order, you cannot refer to an item by using an index.
*  The keys in a dictionary cannot be duplicated

```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(Car)
```

Dictionary can have values of multiple data types. This includes String, int,float , boolean ,list and so on 

```
Car = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(Car)
```

## Accessing elements within a  dictionary


We can use the key to access elements in a dictionary

```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "colors": ["red", "white", "blue"]

}
print("1",Car["model"])
print("2",Car.get("colors"))
```

If we are in need of extracting all the keys within the dictionary, we can use the keys() method. Similarly, if we want to get the values in the dictionary, we can use the values() method.

```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "colors": ["red", "white", "blue"]

}
print(Car.keys())
print(Car.values())
```

If we want to access the keys or the values individually, we can use a for loop to loop through the elements


```
Cars = {
  "Car1": "Ford",
  "Car2": "Nissan",
  "Car3": "Suzuki",
  "Car4": "Mitsubishi",
  "Car5": "Tata",
}

print("This will give the keys within the dictionary")
print()

for x in Cars:
  print(x)

print()
print("This will give the values within the dictionary")
print()
for x in Cars:
  print(Cars[x])

print()
print("This will give the key,values pairs within the dictionary")
print()
for x in Cars:
  print("The brand of ",x,"is",Cars[x])

print()
print("Another way")
print()
for x,y in Cars.items():
  print("The brand of ",x,"is",y)
```

## Updating elements in a dictionary


* There are 2 ways to update elements in a python dictionary


1.   Using the update method
2.   Accessing the element using the key and assigning the new value






```
Car1 = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print("---------------Using the key---------------")
print("Before updating:",Car1)
Car1["year"] = 2018
print("After updating:",Car1)

Car2 = {
  "brand": "Nissan",
  "model": "GTR - R32",
  "year": 2000
}
print()
print("---------------Using the update method---------------")
print("Before updating:",Car2)
Car2.update({"year": 2010})
print("After updating:",Car2)
```

## Adding items to a dictionary

*   This can also be done in 2 ways,


1.   Using the update method in dictionary
2.   By adding a new index key and assigning a value to it







```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print("---------------Using the key---------------")
print("Before adding the color of the car:",Car)
Car["color"] = "red"
print("After adding the color of the car:",Car)

print()
print("---------------Using the update method---------------")
print("Before adding the mileage:",Car)
Car.update({"mileage": "10450km"})
print("After adding the mileage:",Car)
```

## Removing an item / Deleting the dictionary


*   There are 2 methods to delete an item from a python dictionay


1.   pop() method
2.   del 


```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "color":["Red","Blue"]
}
print("---------------Using the pop() method---------------")

print(Car)
Car.pop("year")
print(Car)

print()
print("---------------Using del---------------")
print(Car)
del Car["color"]
print(Car)
```

We can also delete the whole dictionary using the del method


```
Car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "color":["Red","Blue"]
}
print(Car)
del Car
print(Car) # an error will occour since it is deleted
```

#<u> Sets </u>


*   Set is the 4th data type that we will study to store data
*   These are un ordered(meaning that we cannot access using an index and that it doesnt keep the input order) and does not allow duplicate values
*   Sets are defined within 2 curley braces seperated by commas
* Sets are also immutable meaning that we cannot change the items after the set has been created.
*   But we can add or delete items and do operations on the set iteself











```
 basket = {'apple', 'orange', 'apple', 'pear', 'orange','banana'}
print(basket) # notice that there is only one entry of each item
```

We can use a set to remove the duplicate elements from a list, using the set() function in python that builds a set using the inpuut

```
list1 = ['apple','apple','orange']
set1 = set(list1)
print("The list with duplicate elements:",list1)
print("The set without duplicate elements:",set1)
list1 = list(set1)
print("The list without duplicate elements:",list1)
```

## Adding elements to the set
We can use the add() method to add elements to a set

```
fruits = {"apple", "banana", "cherry","apple"}
print("fruits before adding the new value:",fruits)
fruits.add("orange")
print("fruits after adding the new value:",fruits)
```

We can also add elements that are already in a list or a set




```
fruits = {"apple", "cherry","apple"}
tropicalFruits = { "banana","pineapple"}
healthyFruits = ["papaya","dragonfruit"]

print("fruits before adding :",fruits)
fruits.update(tropicalFruits)
print("fruits after adding tropical fruits:",fruits)
fruits.update(healthyFruits)
print("fruits after adding healthy fruits:",fruits)

```

## Removing / discarding elements
*   Removing and discarding achieves the common task of deleting a specific element.
*   However, the difference is, if you remove an element that does not exist within a set, it will throw a runtime exeption.
*   If you discard a non existing element from a set , there would be no effect.





```
fruits = {"apple", "cherry","apple","banana","peach"}
print("fruits before deleting :",fruits)
fruits.remove("apple")
print("fruits after removing :",fruits)
fruits.discard("cherry")
print("fruits after discarding :",fruits)
fruits.discard("mango")
print("fruits after discarding a non existing element:",fruits)



```

## Set operations


*  The possible operations on a set include,

1.   Set difference
2.   Difference update
1.   Intersection
2.   Union
1.   isDisjoint
2.   isSubset

### Set difference
difference()- this method returns the difference of two or more set as a
new set

```
x={"a","b","c","d","e"}
y={"b","c"}
z={"c","d"}
a1 = x.difference(y)
print("x-y:",a1)
a2 = x.difference(y).difference(z)
print("x-y-z:",a2)

print("x:",x)
```

### Difference Update
The method difference_update removes all elements of another set
from this set.

```
x={"a","b","c","d","e"}
y={"b","c"}
x.difference_update(y)
print("x-y:",a1)

print("x:",x)
```

### Intersection
intersection(s) : Returns the intersection of the instance set and the set s as a new set

```
x={"a","b","c","d","e"}
y={"c","d","e","f","g"}

print("x.intersection(y):",x.intersection(y))
print("x & y:",x & y)
```

### Union
Union(s) : Returns the union of the instance set and the set s as a new set

```
A = {1,2,3,4,5}
B = {4,5,6,7,8}
print("A.union(B):",A.union(B))
print("A | B:",A | B)
```

### isDisjoint


*   This returns whether there is an intersection within 2 sets




```
a={1,2,3}
b={4,5,6}
c={2}
print(a.isdisjoint(b))
print(a.isdisjoint(c))
```

### isSubset
This return whther one set is a subset of another or not

```
x={1,2,3,4}
y={1,2,3,4,5,6,7,8,9}
print(x.issubset(y))
```
